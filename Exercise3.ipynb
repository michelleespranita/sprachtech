<div class="task_description">
   <i class="subtask">3.1.3</i>
</div>

For the word(s) with the greatest number of distinct tags, print the sentences from the corpus containing the word.

```python
import string
import re
cfd = nltk.ConditionalFreqDist((word,tag) for (word, tag) in brown.tagged_words(tagset='universal'))
homographs = [(key, len(cfd[key])) for key in cfd.keys() if len(cfd[key])>4]
words = [key for key in cfd.keys() if len(cfd[key])>4]
print("Word(s) with the greatest number of distinct tags:", homographs)
print("List of sentences containing the word: ")
txt = brown.sents()
for sentence in txt:
    if list(set(words) & set(sentence)):
            result = "".join([" "+i if not i.startswith("'") and i not in string.punctuation else i for i in sentence]).strip()
            print(result)
```

What is the ratio of masculine to feminine pronouns?

```python
import math
masculine = [word for word, pos in pos_tag(text) if word.lower() == 'he' or word.lower() == 'his' or word.lower() == 'him' or word.lower() == 'himself']
feminine = [word for word, pos in pos_tag(text) if word.lower() == 'she' or word.lower() == 'her' or word.lower() == 'hers'or word.lower() == 'herself']
x = math.gcd(len(masculine), len(feminine))
print("Ratio of masculine pronouns:feminine pronouns")
print(int(len(masculine)/x),':', int(len(feminine)/x))
```

Print a table with the integers 1..10 in one column, and the number of types in the corpus having 1..10 distinct tags in the other column.

```python
cfd = nltk.ConditionalFreqDist((word,tag) for (word, tag) in brown.tagged_words(tagset='universal'))
i = 1
print("No. of distinct tag(s)\tNo. of types")
while i < 11 :
    homographs = [(key, len(cfd[key])) for key in cfd.keys() if len(cfd[key])==i]
    i += 1
    print(i-1, "\t\t\t", len(homographs))
```
